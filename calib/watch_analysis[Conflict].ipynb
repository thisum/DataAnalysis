```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = open("../gear_user_data/Yilei_raw_data.txt","r")
lines = file.readlines()
data = {}
for line in lines:
    [l,d] = line.split(":")
    d = d.replace('[', '').replace(']', '').replace('\n', '')
    d = np.fromstring(d, dtype=int, sep=',')
    data.update({l:d})
```

```python
categories = []
for key, value in data.items():
    [a,g,r] = key.split("-")
    categories.append(a + "-" + g)
```

```python
# pd.Series(categories).value_counts()
```

```python
N = 5
weights = np.ones(N) / N
# plt.figure(num=1, figsize=(16, 16))

for cat in categories:
    for key, value in data.items():
        if cat in key:
            avg = np.convolve(value, weights, mode='valid')
            x = np.arange(value.size)
            grad = np.gradient(value, x)
            plt.title(cat)
            plt.plot(grad)
            
    plt.show()
```
