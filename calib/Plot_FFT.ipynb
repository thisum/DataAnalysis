```python
import numpy as np
import pandas as pd
from scipy import pi
import matplotlib.pyplot as plt
from scipy import fftpack
from scipy import hamming
from scipy import signal
%matplotlib inline
```

```python
file = "test_ppg_6"
data = pd.read_csv("../gear_data/transitions/"+file+".csv")
```

```python
f, t, Sxx = signal.spectrogram(data.A, fs=10, nperseg=8, noverlap=7, mode='psd')

plt.figure(num=1, figsize=(16, 8))
plt.subplot(141)
plt.pcolormesh(t, f, Sxx)
plt.title("8 frames")
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

f, t, Sxx = signal.spectrogram(data.A, fs=10, nperseg=16, noverlap=15, mode='psd')
plt.subplot(142)
plt.pcolormesh(t, f, Sxx)
plt.title("16 frames")
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

f, t, Sxx = signal.spectrogram(data.A, fs=10, nperseg=32, noverlap=31, mode='psd')
plt.subplot(143)
plt.pcolormesh(t, f, Sxx)
plt.title("32 frames")
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

f, t, Sxx = signal.spectrogram(data.A, fs=10, nperseg=64, noverlap=63, mode='psd')
plt.subplot(144)
plt.pcolormesh(t, f, Sxx)
plt.title("64 frames")
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.savefig(file+".pdf", quality=100)
plt.show()
```
